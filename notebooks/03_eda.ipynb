# Notebook 03 — EDA and Chart Generation
Loads `data/processed/featured.csv`, calls all 12 `plot_*` functions, produces 16 PNG files in `charts/`.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import pandas as pd
from src.visualisations import (
    plot_tier_comparison, plot_city_month_heatmap, plot_monsoon_dip,
    plot_yoy_trend, plot_industrial_corridor, plot_pollutant_correlation,
    plot_diwali_spike, plot_pollutant_dominance, plot_feature_importance_comparison,
    plot_feature_importance_city, plot_model_comparison, plot_industrial_vs_urban
)

FEATURED_CSV = '../data/processed/featured.csv'
CHARTS_DIR   = '../charts'

if not os.path.exists(FEATURED_CSV):
    raise FileNotFoundError(f'featured.csv not found at {FEATURED_CSV}. Run notebook 02 first.')

df = pd.read_csv(FEATURED_CSV, parse_dates=['date'])
print(f'Loaded {len(df):,} rows')
os.makedirs(CHARTS_DIR, exist_ok=True)
```

```python
# 1. Tier comparison
p = plot_tier_comparison(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```

```python
# 2. City-month heatmap
p = plot_city_month_heatmap(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```

```python
# 3. Monsoon dip
p = plot_monsoon_dip(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```

```python
# 4. YoY trend
p = plot_yoy_trend(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```

```python
# 5. Industrial corridor
p = plot_industrial_corridor(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```

```python
# 6. Pollutant correlation
p = plot_pollutant_correlation(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```

```python
# 7. Diwali spike
p = plot_diwali_spike(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```

```python
# 8. Pollutant dominance
p = plot_pollutant_dominance(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```

```python
# 9. Industrial vs urban box plot
p = plot_industrial_vs_urban(df, out_dir=CHARTS_DIR)
print('Saved:', p)
```
